```python
import pandas as pd
import pickle

df = pd.read_csv('../data/test.csv', index_col=0)
```

```python
df.head()
```

```python
X_test = df[['Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']]
X_test_new = X_test[['Embarked', 'Pclass', 'Sex', 'Age']]
```

```python
# load the transformation from disk
fe = pickle.load(open('feature_1.pkl', 'rb'))
```

```python
Xtest_tran = fe.transform(X_test_new)
pd.DataFrame(Xtest_tran, columns=fe.get_feature_names_out())
```

```python
# load the model from disk
model = pickle.load(open('model_1.pkl', 'rb'))
```

```python
predictions = model.predict(Xtest_tran)
predictions
```

```python
X_test.head()
```

```python
import numpy as np

ids = X_test.index
submission = pd.DataFrame({ 'PassengerId' : np.array(ids), 'Survived': np.array(predictions) })
submission.to_csv('submission.csv', index=False)

ids
```
